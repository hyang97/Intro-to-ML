```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean_0 = (0, 0)
cov_identity = [[1,0],[0,1]]
x1, x2 = np.random.multivariate_normal(mean_0, cov_identity, 1000).T
plt.plot(x1, x2)
```

```python
mean_1 = (1, 1)
```

```python
x1, x2 = np.random.multivariate_normal(mean_1, cov_identity, 1000).T
plt.plot(x1, x2)
```

```python
cov_doubled = [[2,0],[0,2]]
x1, x2 = np.random.multivariate_normal(mean_0, cov_doubled, 1000).T
plt.plot(x1, x2)
```

```python
cov_d = [[1, 0.5],[0.5, 1]]
x1, x2 = np.random.multivariate_normal(mean_0, cov_d, 1000).T
plt.plot(x1, x2)
```

```python
cov_e = [[1, -0.5], [-0.5, 1]]
x1, x2 = np.random.multivariate_normal(mean_0, cov_e, 1000).T
plt.plot(x1, x2)
```
